### Generate Clinical Test Std data
Columns:
 _id,	lab_name,	test_key,	test_name,	attribute_key,	attribute_name,	result_type,	loinc_code,	loinc_name,	component,	short_name,	class_name,	unit(UCUM)

```python
# print(s1)
def get_type(test_key, attr_key):
    cat = s1[(s1['test_key']==test_key) & (s1['attr_key']==attr_key)]
#     print(cat['data_type'].values[0])
    return cat['data_type'].values[0]
```

```python
# del std_attr['@']
std_attr['_id'] = std_attr.index + 1001
# print(std_attr)
std_attr['result_type'] = std_attr.apply(lambda x: get_type(x.test_key, x.attribute_key), axis=1)
std_attr
```

### Generate Test Name (All possible)
Columns:	_id,	test_key,	name,	ref,	ref_code

```python
del std_labtest["@"]
```

```python
std_labtest
```

```python
test_name = std_attr[['test_key', 'test_name']]
test_name = test_name.drop_duplicates(subset=['test_key', 'test_name'])
test_name.apply(lambda x: get_testname_syno(x.test_key), axis=1)
```

```python
test_keys = []
name = []
ref = []
ref_code = []
def get_testname_syno(test_key):
#     print(test_key)
    row = std_labtest[std_labtest['test_key']==test_key]
    snomed_ct = row['snomed_name'].values[0].split(',')
    snomed_id = row['snomed_ct_id'].values[0]
    for s in snomed_ct:
        test_keys.append(test_key)
        name.append(s.strip())
        ref.append('SNOMED CT')
        ref_code.append(snomed_id)
    # Icddrb term
    test_keys.append(test_key)
    name.append(row['iccdrb_name'].values[0])
    ref.append('Icddrb')
    ref_code.append(None)
    ## source BSH term
    s1_row = s1[s1['test_key']==test_key]
#     print(s1_row)
    test_keys.append(test_key)
    name.append(s1_row['test_name'].values[0])
    ref.append('Bangladesh Specialized Hospital')
    ref_code.append(None)
    ## source Eye hospital term
    s2_row = s2[s2['test_key']==test_key]
    if(not s2_row.empty):
        test_keys.append(test_key)
        name.append(s2_row['test_name'].values[0])
        ref.append('Bangladesh Eye Hospital')
        ref_code.append(None)
```

```python
name
```

```python
test_name_df = pd.DataFrame(
    {'_id': list(range(1, len(test_keys)+1)),
     'test_key': test_keys,
     'name': name,
     'ref': ref,
     'ref_code': ref_code
    })
test_name_df
```

```python
test_name_df.to_csv(data_dir + 'std/testname.csv', index=False)
```
